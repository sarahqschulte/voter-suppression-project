```python
voter_df = pd.read_excel('../voter-suppression-project/2018_General_Voter_Data.xlsx')
voter_df.head()
```

```python
voter_df = voter_df.rename(columns = {"Unnamed: 6":"2018 Vote for Highest Office", "Denominators":"Voting-Eligible_Population(VEP)","Unnamed: 8":"Voting-Age_Population(VAP)"})
voter_df.head()
```

```python
voter_df = voter_df[['State','2018 Vote for Highest Office','Voting-Eligible_Population(VEP)','Voting-Age_Population(VAP)' ]]
voter_df.head()
```

```python
voter_df = voter_df.drop(0, axis=0)
voter_df.head()
```

```python
voter_df
```

```python
voter_df = voter_df.drop(53, axis = 0)
voter_df = voter_df.drop(54, axis = 0)
voter_df = voter_df.drop(55, axis = 0)
voter_df = voter_df.drop(56, axis = 0)
voter_df = voter_df.drop(57, axis = 0)
voter_df
```

```python
voter_df.dtypes
```

```python
voter_df['2018 Vote for Highest Office'] = voter_df['2018 Vote for Highest Office'].astype(int)
voter_df.dtypes
```

```python
voter_df['Voting-Eligible_Population(VEP)'] = voter_df['Voting-Eligible_Population(VEP)'].astype(int)
voter_df.dtypes
```

```python
voter_df['Voting-Age_Population(VAP)'] = voter_df['Voting-Age_Population(VAP)'].astype(int)
voter_df.dtypes
```

```python
voter_df.head()
```

```python
voter_df['Percentage_Eligible_Voting'] = voter_df['2018 Vote for Highest Office']/voter_df['Voting-Eligible_Population(VEP)']*100
voter_df.head()
```

```python
voter_df['18+_Uneligible'] = voter_df['Voting-Age_Population(VAP)'] - voter_df['Voting-Eligible_Population(VEP)']
voter_df.head()
```

```python
voter_df['18+_Uneligible_Percent'] = voter_df['18+_Uneligible']/voter_df['Voting-Age_Population(VAP)']*100
```
